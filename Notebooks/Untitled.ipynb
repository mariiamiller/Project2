```python
new_list = []
for i in range(length):

    new_list.append(stock_list[2*i])

new_list
```

```python
new_list2 = []
for i in range(length):

    new_list2.append(stock_list[2*i+1])

new_list2
```

```python
df2 = pd.DataFrame(new_list2)
df2.head
```
