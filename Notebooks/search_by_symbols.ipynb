```python
import requests
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

earnings = pd.read_csv('earnings_latest.csv')
ern = earnings[['symbol','date','eps']]
df = pd.DataFrame(ern)
df.sort_values('eps')
sym = df['symbol'].value_counts().index.tolist()
df.set_index('symbol',inplace=True)
```

```python
symbols_dict = dict()
for symbol in sym:
    symbols_dict[symbol] = df.loc[symbol,:]
```
