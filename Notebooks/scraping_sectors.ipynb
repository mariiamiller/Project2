```python
df = pd.read_csv('earnings_latest.csv')
df
```

```python
symbols = df['symbol'].unique()
symbols
```

```python
sectors = []
for symbol in symbols:
    url = 'https://finance.yahoo.com/quote/'+symbol+'/profile?p='+symbol
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    #browser.click_link_by_partial_text('next')
    sector_list = []
    results = soup.find_all('span', class_="Fw(600)")
    sector_list.append(symbol)

    
    if (results):
        sector = results[0].text
        industry = results[1].text
        
        sector_list.append(sector)
        sector_list.append(industry)
        sectors.append(sector_list)
    else:
        sector_list.append('NULL')
        sector_list.append('NULL')
        sectors.append(sector_list)
    print(results)
```

```python
sectors = []
for symbol in symbols:
    url = 'https://finance.yahoo.com/quote/'+symbol+'/profile?p='+symbol
    response = requests.get(url)
    soup = BeautifulSoup(response.text)

    #browser.click_link_by_partial_text('next')
    sector_list = []
    results = soup.find_all('span', class_="Fw(600)")
    sector_list.append(symbol)

    
    if (results):
        sector = results[0].text
        industry = results[1].text
        
        sector_list.append(sector)
        sector_list.append(industry)
        sectors.append(sector_list)
    else:
        sector_list.append('NULL')
        sector_list.append('NULL')
        sectors.append(sector_list)
    print(results)
```

```python
df_sectors = pd.DataFrame(sectors, columns = ['symbol','sector','industry'])
df_sectors
```
