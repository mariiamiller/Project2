```python
symbol_list = df['symbol'].unique()
symbol_list
```

```python
surp_score = []
beat_score = []
for symbol in symbol_list:
    df_a= df[df["symbol"] == symbol]
    df_a['sqr']=df['surp']*df['surp']
    sum_sqr=df_a['sqr'].sum()
    mult = len(df_a)*(len(df_a)-1)
    surp_score.append(math.sqrt(sum_sqr/mult)*stats.t.ppf(1-0.025, len(df_a)))
    beat = df_a['surp'][df_a['surp']>=0].count()
    beat_score.append(beat/len(df_a))
    print(symbol)
    
    
```

```python
surp_score
```

```python
surp_summ= pd.DataFrame(list(zip(symbol_list, surp_score, beat_score)), 
               columns =['symbol', 'surp_score','beat_score'])
surp_summ
```

```python
surp_summ.to_csv('surprise_summary.csv')
```
